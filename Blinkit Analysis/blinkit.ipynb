```python
import pandas as pd
import sqlite3

file_path = 'BlinkIT Grocery Data.csv'
df = pd.read_csv(file_path)

connection = sqlite3.connect('my_database.db')

df.to_sql('blinkit', connection, if_exists='replace', index=False)

# Step 4: Query the SQLite database to check if the data is loaded
cursor = connection.cursor()
cursor.execute("SELECT * FROM blinkit")

# Fetch and display the results
rows = cursor.fetchall()
for row in rows:
    print(row)


```

```python
column_names = df.columns.tolist()

print("Column names:", column_names)
```

```python
average_rating = df['Rating'].mean()
print(f"\nAverage Rating: {average_rating:.2f}")
```

```python
average_sales = df['Sales'].mean()
print(f"Average Sales: {average_sales:.2f}")
```

```python
total_sales = df['Sales'].sum()
print(f"Total Sales: {total_sales:.2f}")
```

```python
low_fat_count = df[df['Item Fat Content'] == 'Low Fat'].shape[0]
regular_count = df[df['Item Fat Content'] == 'Regular'].shape[0]

print(f"\nCount of Low Fat Items: {low_fat_count}")
print(f"Count of Regular Items: {regular_count}")
```

```python
average_sales_low_fat = df[df['Item Fat Content'] == 'Low Fat']['Sales'].mean()
average_sales_regular = df[df['Item Fat Content'] == 'Regular']['Sales'].mean()

print(f"\nAverage Sales for Low Fat Items: {average_sales_low_fat:.2f}")
print(f"Average Sales for Regular Items: {average_sales_regular:.2f}")
```

```python

average_rating_low_fat = df[df['Item Fat Content'] == 'Low Fat']['Rating'].mean()
average_rating_regular = df[df['Item Fat Content'] == 'Regular']['Rating'].mean()

print(f"\nAverage Rating for Low Fat Items: {average_rating_low_fat:.2f}")
print(f"Average Rating for Regular Items: {average_rating_regular:.2f}")
```

```python
item_type_counts = df['Item Type'].value_counts().reset_index()
item_type_counts.columns = ['Item Type', 'Count']
print("\nCount of Items by Type:")
print(item_type_counts)
```

```python
average_sales_by_outlet = df.groupby('Outlet Type')['Sales'].mean().reset_index()
average_sales_by_outlet.columns = ['Outlet Type', 'Average Sales']
print("\nAverage Sales by Outlet Type:")
print(average_sales_by_outlet)
```

```python
total_sales_by_location = df.groupby('Outlet Location Type')['Sales'].sum().reset_index()
total_sales_by_location.columns = ['Outlet Location Type', 'Total Sales']
print("\nTotal Sales by Outlet Location Type:")
print(total_sales_by_location)
```

```python
correlation = df['Sales'].corr(df['Rating'])
print(f"\nCorrelation Between Sales and Ratings: {correlation:.2f}")
```
