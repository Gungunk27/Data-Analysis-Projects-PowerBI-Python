```python
import pandas as pd

# Load the Excel file
excel_file = 'Adidas US Sales Datasets.xlsx'  # Replace with your Excel file path
df = pd.read_excel(excel_file)

csv_file = 'converted_file.csv'
df.to_csv(csv_file, index=False)
```

```python
# Load the CSV file into a DataFrame
data = pd.read_csv(csv_file)
data
```

```python
total_sales_per_retailer = data.groupby('Retailer')['Total Sales'].sum().reset_index().sort_values(by='Total Sales', ascending=False)
print(total_sales_per_retailer)
```

```python
top_products = data.groupby('Product')['Total Sales'].sum().reset_index().sort_values(by='Total Sales', ascending=False).head(10)
print(top_products)
```

```python
sales_by_region = data.groupby('Region')['Total Sales'].sum().reset_index().sort_values(by='Total Sales', ascending=False)
print(sales_by_region)
```

```python
average_price_per_product = data.groupby('Product')['Price per Unit'].mean().reset_index().sort_values(by='Price per Unit', ascending=False)
print(average_price_per_product)
```

```python
units_sold_by_method = data.groupby('Sales Method')['Units Sold'].sum().reset_index().sort_values(by='Units Sold', ascending=False)
print(units_sold_by_method)
```

```python
top_states_profit = data.groupby('State')['Operating Profit'].sum().reset_index().sort_values(by='Operating Profit', ascending=False).head(5)
print(top_states_profit)
```

```python
data['Invoice Date'] = pd.to_datetime(data['Invoice Date'])
monthly_sales = data.groupby(data['Invoice Date'].dt.month)['Total Sales'].sum().reset_index().sort_values(by='Invoice Date')
print(monthly_sales)
```

```python
average_operating_margin = data.groupby('Retailer')['Operating Margin'].mean().reset_index().sort_values(by='Operating Margin', ascending=False)
print(average_operating_margin)
```

```python
city_product_sales = data.groupby(['City', 'Product'])['Units Sold'].sum().reset_index().sort_values(by='Units Sold', ascending=False)
print(city_product_sales)
```

```python
sales_performance = data.groupby(['Retailer', 'Region'])['Total Sales'].sum().reset_index().sort_values(by='Total Sales', ascending=False)
print(sales_performance)
```
